```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentense = "Natural Language Processing helps computers to understand human"
```

```python
tokens = word_tokenize(sentense)
```

```python
tags = pos_tag(tokens)
print("Pos Tags:", tags)
```
